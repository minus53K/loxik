# 2. 使用方法

现在，假如我们已知命题 $a$ 和 $a\Rightarrow b$ 是成立的，我们将它们加入定理集，我们希望得到的结论是 $b$ 成立，现在有两种验证方法：语义和语法

```python
# 导入关键函数
import sys
sys.path.insert(0, '.')
from src.logic_checker import semanticEntailment, syntacticProofCheck

# 引入公理集
theorems = ("a", "(~(a&(~b)))")  # a 和 a->b
```

## 语义

语义层面的验证方法是通过穷举所有命题变量可能取到的真值组合（称为赋值），并检查是否每一个使定理集为真的赋值都能让结论为真

```python
conclusion = "b"
print(semanticEntailment(theorems, conclusion))
```

如果缺少条件，就不能得出语义蕴含

```python
print(semanticEntailment(("a"), conclusion))
```

## 语法

语法层面的验证方法是检查已知的证明序列是否是合法的，也就是说，证明序列是否只出现了给定的定理和逻辑公理，以及是否按照 MP 得到结论。对于上面的例子，我们用一个正确的证明和一个错误的证明来说明

```python
proof = ("a", "(~(a&(~b)))", "b")
print(syntacticProofCheck(theorems, proof))

proof = ("a", "b")
print(syntacticProofCheck(theorems, proof))
```
